```python
import pandas as pede
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

df = pede.read_csv("combined_data.csv", on_bad_lines='skip', encoding='latin-1')

df.rename(columns={df.columns[2]: 'Tweet'}, inplace=True)
df['Tweet'] = df['Tweet'].astype(str)
def label_sementara(text):
    if any(x in text.lower() for x in ['rusak', 'gangguan', 'telat', 'lama']): return 0 
    return 1 
df['Sentimen'] = df['Tweet'].apply(label_sementara)
x = df['Tweet']
y = df['Sentimen']

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42) 

scaler = TfidfVectorizer(max_features=1000)
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)

K = 5
model = KNeighborsClassifier(n_neighbors = K)
model.fit(x_train, y_train)

y_pred = model.predict(x_test)

accuracy = accuracy_score(y_test, y_pred)
print('Metode klasifikasi (KNN pada Teks)\n')
print(f"Akurasi = {accuracy}")
print("\nLaporan Klasifikasi : ")
print(classification_report(y_test, y_pred))

print("\nMatriks Konfusi : ")
print(confusion_matrix(y_test, y_pred))

pasien_baru = ["LRT Jabodebek hari ini gangguan terus dan pintunya rusak"] 
skala_pasien_baru = scaler.transform(pasien_baru)
prediksi = model.predict(skala_pasien_baru)

if prediksi[0] == 0 :
   print("\nTweet Baru Terindikasi Sentimen Negatif (0).")
else :
   print("\nTweet Baru Terindikasi Sentimen Positif (1).")
```
